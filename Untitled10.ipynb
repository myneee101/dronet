```python
dataset = pd.read_csv(r'audi_2022.csv')
#print(df.head(20))
#print(df.shape)
#print(df.describe())
#col = ['model', 'year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']
encod_col = ['model', 'year', 'transmission', 'fuelType']

## checking number of labels each variables has
for col in dataset.columns:
    print(col, ': ', len(dataset[col].unique()), 'labels')
```

```python

print(dataset.describe(include= "all"))

#print(dataset.groupby('model').size())

to_replace1 = [' A1',' A2',' A3',' A4',' A5',' A6',' A7',' A8',' Q2',' Q3',' Q5',' Q7',' Q8',' R8',' RS3',' RS4',' RS5',' RS6',' RS7',' S3',' S4',' S5',' S8',' SQ5',' SQ7',' TT']

to_replace2 = ['Automatic', 'Semi-Auto', 'Manual']

to_replace3 = ['Diesel', 'Petrol','Hybrid']


r = dataset.shape

_value = ['model','transmission','fuelType']
```

```python
##########      Creating cLasses        ###############

nn = 0

for j in to_replace2:
    dataset['transmission'] = dataset['transmission'].replace(j,nn)
    nn += 1

nm = 0
for k in to_replace1:
    dataset['model'] = dataset['model'].replace(str(k),nm)
    nm += 1

no = 0
for l in to_replace3:
    dataset['fuelType'] = dataset['fuelType'].replace(l,no)
    no += 1

print(dataset.describe(include= "all"))
print(dataset.groupby('transmission').size())
print(dataset.groupby('model').size())
print(dataset.groupby('fuelType').size())
```

```python
Matrix_A = dataset.head(10)

# SVD
U, s, VT = svd(Matrix_A)
print("U", U)
print("s", s)
print("VT", VT)
```

```python
# create m x n Sigma matrix
A_or = Matrix_A
Sigma = zeros((A_or.shape[0], A_or.shape[1]))
# populate Sigma with n x n diagonal matrix
Sigma[:A_or.shape[1], :A_or.shape[1]] = diag(s)
```

```python
# reconstruct matrix
Reconstructed_A = U.dot(Sigma.dot(VT))
print("Reconstructed_A")
print(Reconstructed_A)
```

```python
n_elements = 3
Sigma = Sigma[:, :n_elements]
VT = VT[:n_elements, :]

# transform
Transformed_A = U.dot(Sigma)
print("Transformed_A")
print(Transformed_A)

Transformed_A_Re = U.dot(Sigma.dot(VT))
print("Transformed_A_Re")
print(Transformed_A_Re)
```
