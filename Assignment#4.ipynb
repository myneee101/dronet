# 4)	Generate a Random_interget_data of size (10,5) between 1 to 10 values;

# a.	Perform the SVD of the Random_interget_data ? Paste your Code

```python
import numpy as np
from scipy.linalg import svd
Y= np.random.randint(1,10, size=(10,5))
print(Y)
U,s,VT=svd(Y)
print(U)
print(s)
print(VT)
```

# b.	Reading:-- https://machinelearningmastery.com/singular-value-decomposition-for-machine-learning/

# c.	What are the sizes of the U,s,VT ?


```python
print(U.shape)
print(s.shape)
print(VT.shape)
```

# d.	Reconstruct the Random_interget_data by using the U,S and VT

```python
Sigma = np.zeros((Y.shape[0], Y.shape[1]))
Sigma[:Y.shape[1], :Y.shape[1]] = np.diag(s)
X = U.dot(Sigma.dot(VT))
print(X)
```

# e.	What is the reduced feature transform of  Random_interget_data

```python
n_elements = 3
Sigma = Sigma[:, :n_elements]
VT = VT[:n_elements, :]
T = U.dot(Sigma)
print(T)
```

```python
T = Y.dot(VT.T)
print(T)
```

# f.	Reconstruct the Random_interget_data by using the reduced dimension of SVD transformations (only 3 out of 5) in U,S and VT

```python
Z = U.dot(Sigma.dot(VT))
print(Z)
print(np.round(Z))
```

```python
import pandas as pd
from sklearn.svm import SVC 
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
dataset= pd.read_csv("seeds_dataset.csv")
array = dataset.values 
X = array[:,0:7] 
Y = array[:,7]
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=0.20, 
random_state=1)
models = []
models.append(('SVM', SVC(gamma='auto')))
results = []
names = []
for name, model in models:
    kfold = StratifiedKFold(n_splits=10, random_state=1, shuffle=True) 
    cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))
```
